### Simple Logistic Regression Model

Let's implement a simple logistic regression model using `scikit-learn` to classify the data. We'll use the one-variable data `x_train` and `y_train`.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Reshape x_train for scikit-learn (needs to be 2D)
X_train_reshaped = x_train.reshape(-1, 1)

# Create a logistic regression model
model = LogisticRegression()

# Train the model
model.fit(X_train_reshaped, y_train)

# Make predictions on the training data
y_pred = model.predict(X_train_reshaped)

# Calculate accuracy
accuracy = accuracy_score(y_train, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")

# Get the model's parameters
w = model.coef_[0][0]
b = model.intercept_[0]
print(f"Model parameters: w={w:.2f}, b={b:.2f}")

# Plot the data and the decision boundary
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.scatter(x_train[y_train == 1], y_train[y_train == 1], marker='x', s=80, c='red', label="y=1")
ax.scatter(x_train[y_train == 0], y_train[y_train == 0], marker='o', s=100, label="y=0", facecolors='none', edgecolors=dlc["dlblue"], lw=3)

# Plot the sigmoid function (model's prediction probability)
x_plot = np.linspace(x_train.min() - 1, x_train.max() + 1, 100).reshape(-1, 1)
# The probability is given by 1 / (1 + exp(-(w*x + b)))
z = model.decision_function(x_plot) # This gives w*x + b
prob = 1 / (1 + np.exp(-z))
ax.plot(x_plot, prob, color='green', label='Logistic Regression (Prob)')

# Plot the decision boundary (where probability is 0.5)
decision_boundary_x = -b / w
ax.axvline(x=decision_boundary_x, color='purple', linestyle='--', label='Decision Boundary (0.5)')

ax.set_ylim(-0.08, 1.1)
ax.set_ylabel('y', fontsize=12)
ax.set_xlabel('x', fontsize=12)
ax.set_title('Logistic Regression Classification')
ax.legend()
plt.show()
```
